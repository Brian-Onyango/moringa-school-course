#Data Cleaning 

```python
#Changing Column Names on df_CDR1 to be uniform across the three datasets
df_CDR11 = df_CDR1.rename(columns={'PRODUTC': 'PRODUCT','DATETIME' : 'DATE_TIME'})
df_CDR11.head()
# df_CDR22 = df_CDR2.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'})
# df_CDR33 = df_CDR3.rename(columns={'CELLID': 'CELL_ID'})
```

```python
#Changing Column Names on df_CDR2 to be uniform across the three datasets
df_CDR22 = df_CDR2.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'})
df_CDR22.head()
```

```python
#Changing Column Names on df_CDR3 to be uniform across the three datasets
df_CDR33 = df_CDR3.rename(columns={'CELLID': 'CELL_ID','SIET_ID':'SITE_ID'})
df_CDR33.head()
```

```python
#Changing Column Names on df_cells_geo to be uniform across the datasets
df_cells_geo = df_cells_geo.rename(columns={'SITE_CODE':'SITE_ID'})
df_cells_geo.head()
```

```python
#Cleaning the CELL_ID column
df_CDR11['CELL_ID'] = df_CDR11['CELL_ID'].str.replace('+','')
df_CDR11['CELL_ID'] = df_CDR11['CELL_ID'].str.replace('/','')
df_CDR22['CELL_ID'] = df_CDR22['CELL_ID'].str.replace('+','')
df_CDR22['CELL_ID'] = df_CDR22['CELL_ID'].str.replace('/','')
df_CDR33['CELL_ID'] = df_CDR33['CELL_ID'].str.replace('+','')
df_CDR33['CELL_ID'] = df_CDR33['CELL_ID'].str.replace('/','')
```

#Data Preparation

```python
#concatanating the three tables
join = [df_CDR11, df_CDR22, df_CDR33]
result = pd.concat(join)
result.head()
```

```python
#result table info
result.info()
```

```python
#Joining the result table to df_cells_geo on CELL_ID
dataset = pd.merge(result, df_cells_geo, on='SITE_ID', how="left")

dataset.head(200)
```
