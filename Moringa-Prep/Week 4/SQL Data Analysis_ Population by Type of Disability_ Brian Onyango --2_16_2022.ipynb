## 1.5 Answering Questions

Let's answer the following questions with our dataset. These are the same questions we answered using the python programming language. 

```
# Challenge 1
# Which county had the highest no. of registered deaf persons?
%%sql
SELECT COUNTY, DEAF*TOTAL_COUNT FROM POPULATION
ORDER BY DEAF*TOTAL_COUNT DESC LIMIT 1
```

```
# Challenge 2
# Which county had the highest no. of registered persons with a missing hand?
%%sql
SELECT COUNTY, MISSING_HAND*TOTAL_COUNT FROM POPULATION
ORDER BY MISSING_HAND*TOTAL_COUNT DESC LIMIT 1
```

```
# Challenge 3
# Which county had the highest no. of registered persons with a missing foot?
%%sql
SELECT COUNTY, MISSING_FOOT*TOTAL_COUNT FROM POPULATION
ORDER BY MISSING_FOOT*TOTAL_COUNT DESC LIMIT 1
```

```
# Challenge 4
# Which county had the highest no. of registered lame persons?
%%sql
SELECT COUNTY, LAME*TOTAL_COUNT FROM POPULATION
ORDER BY LAME*TOTAL_COUNT DESC LIMIT 1
```

```
# Challenge 5
# Which county had the lowest no. of registered blind persons?
%%sql
SELECT COUNTY, BLIND*TOTAL_COUNT FROM POPULATION
WHERE BLIND*TOTAL_COUNT != 0
ORDER BY BLIND*TOTAL_COUNT ASC LIMIT 1
```

```
# Challenge 6
# Which county had the highest third no. of registered deaf persons?
%%sql
SELECT COUNTY, DEAF*TOTAL_COUNT, RANK(), OVER(ORDER BY DEAF*TOTAL_COUNT DESC) AS rank) FROM POPULATION
```

```
# Challenge 7
# In descending order, which top 5 counties had the highest no. of registered dumb persons?
%%sql
SELECT COUNTY, DUMB*TOTAL_COUNT FROM POPULATION
ORDER BY DUMB*TOTAL_COUNT DESC LIMIT 5
```

```
# Challenge 8
# In ascending order, which top 5 counties had the highest no. of registered persons with a mental disability? 
%%sql
SELECT COUNTY, MENTAL*TOTAL_COUNT FROM POPULATION
ORDER BY MENTAL*TOTAL_COUNT DESC LIMIT (5)
```

```
# Challenge 9
# Which counties had no registerd blind persons nor deaf persons?
%%sql
SELECT COUNTY FROM POPULATION
WHERE DEAF =0 AND BLIND =0
```

```
# Challenge 10
# Which disability was the most registered across all the counties?
%%sql
SELECT COUNT(Mental) FROM POPULATION
WHERE LAME != 0 
```

```
# Challenge 11
# Which disability was the least registered across all the counties?
#
%%sql
SELECT COUNT(Missing_Hand) FROM POPULATION
WHERE Missing_Hand != 0 
```

```
# Challenge 12
# What was the average no. of registered persons with a disability?
%%sql
SELECT AVG(TOTAL_COUNT) FROM POPULATION
```

```
# Challenge 13
# Which three counties had least registered persons with disabilities?
%%sql
SELECT COUNTY,TOTAL_COUNT FROM POPULATION
ORDER BY TOTAL_COUNT ASC LIMIT 3
```

```
# Challenge 14
# What was the total no of registered persons with a disability across all counties?
%%sql
SELECT SUM(TOTAL_COUNT) FROM POPULATION
```

```
# Challenge 15 
# Which top 3 counties has the highest no. of registered persons with a disability?
%%sql
SELECT COUNTY,TOTAL_COUNT FROM POPULATION
ORDER BY TOTAL_COUNT DESC LIMIT 3
```
