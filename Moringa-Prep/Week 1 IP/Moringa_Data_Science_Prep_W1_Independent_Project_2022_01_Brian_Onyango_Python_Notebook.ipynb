```python
#Take in two values from a user (both in pounds) then convert those values to kilograms.
p1=int(input("Enter value 1 in Pound:"))
p2=int(input("Enter value 2 in Pound:"))

val1=p1*0.453592
val2=p2*0.453592

print("Value 1 is",val1,"Kgs and value 2 is",val2,"Kgs")




```

```python
#Perform the sum of the values.
k=(val2,val1)
print(sum(k))
```

```python
#Perform the average of the values.
import statistics
print(statistics.mean(k))
```

```python
#Find the difference between both values.  
import numpy as np
print(np.diff(k))
```

```python
#Find the quotient when one value is divided by the other.
print(val1//val2)
```
