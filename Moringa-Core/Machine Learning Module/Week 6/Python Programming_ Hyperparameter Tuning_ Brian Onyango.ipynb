# Python Programming: Hyperparameter Tuning

## Example 

```
# Loading libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets
from sklearn.model_selection import GridSearchCV
```

```
# Loading data
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```
# Creating the logistic regression
logistic = linear_model.LogisticRegression(solver='lbfgs')
```

```
# Creating regularization penalty space
penalty = ['l1', 'l2']

# Creating regularization hyperparameter space
C = np.logspace(0, 4, 10)

# Creating hyperparameter options
hyperparameters = dict(C=C, penalty=penalty)
```

```
# Creating grid search using 5-fold cross validation
clf = GridSearchCV(logistic, hyperparameters, cv=5, verbose=0)
```

```
# Fitting grid search
best_model = clf.fit(X, y)
```

```
# Viewing best hyperparameters
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

```
# Predicting target vector
best_model.predict(X)
```

## <font color="green">Challenge 1</font>

```
# Challenge 1
# ---
# Create a linear logistic regression model to predict the gender. 
# Perform hyperparameter tuning as well. 
# ---
# Dataset 

url = 'http://bit.ly/HeightsWeightsDataset'
df =pd.read_csv(url)
df.head()
```

```
# Understanding the data
df.info()
```

```
# Understanding Gender colum
df.Gender.unique()
```

```
# Label encoding Gender columns to Binomial for our logistic model
df = df.replace({"Gender":{"Male":1,"Female":0}})
df.head()
```

```
# Splitting our dataset

X = df.drop("Gender",axis=1)
y = df["Gender"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```
# Creating the logistic regression
logistic = linear_model.LogisticRegression()

# Creating regularization penalty space
penalty = ['l1', 'l2']

# Creating regularization hyperparameter space
C = np.logspace(0, 4, 10)

# Creating hyperparameter options
hyperparameters = dict(C=C, penalty=penalty)

# Creating grid search using 5-fold cross validation
clf = GridSearchCV(logistic, hyperparameters, cv=5, verbose=0)
```

```
# Training/Fitting grid search
best_model = clf.fit(X, y)
```

```
# Using our model to make a prediction
#
y_pred = best_model.predict(X_test)
```

```
# comparing the test and predicted
comparison_frame = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
comparison_frame.head(10)
```

```
# describe
comparison_frame.describe()
```

```
# Evaluating the model
#
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy' , accuracy_score(y_test, y_pred))
# The results from the confusion matrix tell us that 1846 out of 2000 are the number of correct predictions. 
# Accuracy score is 92.3%
```

```
from sklearn import metrics

# Our first metric is MAE - Mean absolute error
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))

# We can also use MSE - Mean squared error
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  

# Finally, the most popular metric: RMSE - Root mean squared error
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

mean = df["Gender"].mean()
print(mean)
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
x = 0.1*mean
if rmse >x:
  print("The model/algorithm is not Accurate")
else:
  print("The model/algorithm is Accurate")
```
